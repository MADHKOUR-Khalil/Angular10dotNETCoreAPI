#Transformation de type des colonnes

```python
from pyspark.sql.types import IntegerType,BooleanType,DateType,DecimalType,DoubleType
# Convert String to Integer Type

MSCdf.withColumn("EVENT_DURATION",MSCdf.EVENT_DURATION.cast(IntegerType()))

CCNdf.withColumn("EVENT_DURATION",CCNdf.EVENT_DURATION.cast(DoubleType()))
CCNdf.withColumn("DATA_VOLUME",CCNdf.DATA_VOLUME.cast(DoubleType()))
CCNdf.withColumn("CHARGE_AMOUNT_ORIG",CCNdf.CHARGE_AMOUNT_ORIG.cast(DoubleType()))


MMGdf.withColumn("EVENT_DURATION",MMGdf.EVENT_DURATION.cast(IntegerType()))
MMSdf.withColumn("EVENT_DURATION",MMSdf.EVENT_DURATION.cast(IntegerType()))

PSMdf.withColumn("EVENT_DURATION",PSMdf.EVENT_DURATION.cast(IntegerType()))


MMSdf.withColumn("DATA_VOLUME",MMSdf.DATA_VOLUME.cast(DoubleType()))
```

# Aggregation (Group By, Count, Sum)

```python
MMG_AGG=MMGdf.groupBy("PROC_DATE","PROC_HOUR","EVENT_TYPE","CALL_TYPE","EVENT_STATUS","SUBSCRIBER_TYPE").agg(
    _count("*").alias("CDR_COUNT"),_count("*").alias("MMG_EVENT_COUNT")).collect()
```

```python
MSC_AGG=MSCdf.groupBy("PROC_DATE","PROC_HOUR","EVENT_TYPE","SUBSCRIBER_TYPE","TRAFFIC_TYPE").agg(
    psf.count("*").alias("CDR_COUNT"),psf.sum("EVENT_DURATION").alias("EVENT_DURATION")).collect()

CCN_AGG=CCNdf.groupBy("PROC_DATE","PROC_HOUR","EVENT_TYPE","CALL_TYPE","SUBSCRIBER_TYPE","PARTNER","ROAMING_TYPE","TRAFFIC_TYPE").agg(
    _count("*").alias("CDR_COUNT"),_count("*").alias("CCN_EVENT_COUNT"),_sum("CHARGE_AMOUNT_ORIG").alias("CHARGE_AMOUNT"),psf.sum("DATA_VOLUME").alias("DATA_VOLUME")).collect()

CCN_AGG_msc=CCNdf_msc.groupBy("PROC_DATE","PROC_HOUR","EVENT_TYPE","CALL_TYPE","SUBSCRIBER_TYPE","PARTNER","ROAMING_TYPE","TRAFFIC_TYPE").agg(
    _count("*").alias("CDR_COUNT"),_count("*").alias("CCN_EVENT_COUNT"),_sum("CHARGE_AMOUNT_ORIG").alias("CHARGE_AMOUNT"),psf.sum("DATA_VOLUME").alias("DATA_VOLUME")).collect()

MMS_AGG=MMSdf.groupBy("PROC_DATE","PROC_HOUR","EVENT_TYPE","CALL_TYPE","EVENT_STATUS","TRAFFIC_TYPE","SUBSCRIBER_TYPE","PARTNER").agg(
    _count("*").alias("CDR_COUNT"),psf.sum("DATA_VOLUME").alias("DATA_VOLUME")).collect()

PSM_AGG=PSMdf.groupBy("PROC_DATE","PROC_HOUR","EVENT_TYPE","CALL_TYPE","EVENT_STATUS","SUBSCRIBER_TYPE").agg(
    _count("*").alias("CDR_COUNT"),_count("*").alias("MMG_EVENT_COUNT")).collect()

MMGdf.groupBy("PROC_DATE","PROC_HOUR","EVENT_TYPE","CALL_TYPE","EVENT_STATUS","SUBSCRIBER_TYPE").agg(
    _count("*").alias("CDR_COUNT"),_count("*").alias("MMG_EVENT_COUNT")).collect()
```
